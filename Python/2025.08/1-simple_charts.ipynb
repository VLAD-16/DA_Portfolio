```python
# -*- coding: utf-8 -*-
# Імпорт необхідних бібліотек
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Завантаження даних з файлу
# Припустимо, у вас є файл 'sample_data.csv' у тому ж каталозі
# Якщо файлу немає, можна створити DataFrame вручну для прикладу
try:
    df = pd.read_csv('sample_data.csv')
except FileNotFoundError:
    print("Файл 'sample_data.csv' не знайдено. Створюємо приклад даних.")
    data = {
        'Категорія': ['A', 'B', 'C', 'A', 'B', 'C', 'A', 'B', 'C'],
        'Значення': [10, 15, 20, 12, 18, 25, 11, 16, 22],
        'Дата': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05', '2024-01-06', '2024-01-07', '2024-01-08', '2024-01-09'])
    }
    df = pd.DataFrame(data)

# Виведення перших 5 рядків DataFrame для огляду
print("Перші 5 рядків даних:")
print(df.head())

# Виведення загальної інформації про DataFrame
print("\nІнформація про DataFrame:")
df.info()

# Виведення описової статистики
print("\nОписова статистика:")
print(df.describe())
```

Текстовий блок 1

```python
# Розрахунок середнього значення колонки 'Значення'
average_value = df['Значення'].mean()
print(f"Середнє значення: {average_value:.2f}")
```

```python
# Побудова стовпчастої діаграми, що показує середнє значення для кожної категорії
average_by_category = df.groupby('Категорія')['Значення'].mean().reset_index()

plt.figure(figsize=(8, 6))
plt.bar(average_by_category['Категорія'], average_by_category['Значення'], color='skyblue')
plt.title('Середнє значення за категоріями')
plt.xlabel('Категорія')
plt.ylabel('Середнє значення')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
# Побудова скрипкового графіка для візуалізації розподілу даних за категоріями
plt.figure(figsize=(8, 6))
sns.violinplot(x='Категорія', y='Значення', data=df, palette='pastel')
plt.title('Розподіл значень за категоріями')
plt.xlabel('Категорія')
plt.ylabel('Значення')
plt.show()
```

```python
# Побудова лінійного графіка для візуалізації динаміки даних з часом
plt.figure(figsize=(10, 6))
sns.lineplot(x='Дата', y='Значення', hue='Категорія', data=df, marker='o')
plt.title('Динаміка значень за часом')
plt.xlabel('Дата')
plt.ylabel('Значення')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Цей код будує лінійний графік, що ідеально підходить для відображення змін даних з часом, наприклад, трендів продажів чи температур.
